# Regressão Logística

A Regressão Logística é um algoritmo fundamental no aprendizado de máquina, amplamente utilizado para **problemas de classificação binária**. **Apesar do nome 'regressão', sua finalidade é prever a probabilidade de uma instância pertencer a uma determinada classe**. Apesar do nome “regressão”, ela é usada principalmente para: CLASSIFICAÇÃO. A regressão logística transforma uma combinação linear em uma probabilidade.

## 1. Definição Formal

A Regressão Logística é um modelo linear generalizado que utiliza uma função logística (também conhecida como função sigmoide) para modelar a probabilidade de um resultado binário. Para um dado conjunto de características de entrada $X = (x_1, x_2, ..., x_n)$, o modelo estima a probabilidade $P(Y=1|X)$ através da seguinte equação:

$$P(Y=1|X) = \frac{1}{1 + e^{- (\beta_0 + \beta_1 x_1 + ... + \beta_n x_n)}}$$

Onde:
- $P(Y=1|X)$ é a probabilidade de a variável dependente ser 1 (classe positiva) dadas as características $X$.
- $\beta_0$ é o intercepto.
- $\beta_i$ são os coeficientes de regressão que representam a contribuição de cada característica $x_i$.
- $e$ é a base do logaritmo natural.

A expressão dentro da exponencial, $(\beta_0 + \beta_1 x_1 + ... + \beta_n x_n)$, é a parte linear do modelo, semelhante à regressão linear. A função sigmoide 'comprime' essa saída linear entre 0 e 1, interpretando-a como uma probabilidade.

## 2. Conceitos Fundamentais

- **Variável Dependente Binária**: A saída do modelo deve ser categórica com duas classes (e.g., sim/não, 0/1, compra/não compra).
- **Função Sigmoide (Logística)**: Mapeia qualquer valor real para um valor entre 0 e 1, interpretado como probabilidade.
- **Odds Ratio**: A razão entre a probabilidade de um evento ocorrer e a probabilidade de não ocorrer. A regressão logística modela o logaritmo natural das odds (log-odds ou logit) como uma combinação linear das características de entrada.
- **Estimativa dos Coeficientes**: Os coeficientes ($\beta_i$) são estimados usando o método de Máxima Verossimilhança (Maximum Likelihood Estimation - MLE), que busca os coeficientes que maximizam a probabilidade de observar os dados reais.
- **Função de Custo (Log Loss)**: Também conhecida como entropia cruzada, é a função de custo otimizada durante o treinamento para medir a diferença entre as probabilidades previstas e os rótulos verdadeiros. Para classificação binária, é dada por:
    $$J(\beta) = -\frac{1}{N} \sum_{i=1}^{N} [y_i \log(p_i) + (1 - y_i) \log(1 - p_i)]$$

## 3. Tipos (Variações)

-   **Regressão Logística Binária**: O tipo mais comum, usado quando a variável dependente tem exatamente duas classes.
-   **Regressão Logística Multinomial**: Usada quando a variável dependente tem três ou mais classes nominais (não ordenadas). O modelo prevê a probabilidade de cada classe independentemente.
-   **Regressão Logística Ordinal**: Aplicada quando a variável dependente tem três ou mais classes ordenadas (e.g., baixo, médio, alto). Considera a ordem intrínseca das categorias.

## 4. Macetes para Profissionais de Mercado

1.  **Interpretabilidade é Chave**: A Regressão Logística é altamente interpretável. Os coeficientes podem ser transformados em Odds Ratios ($e^{\beta_i}$), que indicam o quanto a chance de sucesso (Y=1) muda para cada unidade de aumento na característica, mantendo outras constantes.
2.  **Pré-processamento de Dados**:
    *   **Escalonamento**: Embora a Regressão Logística não seja tão sensível a escalonamento quanto SVMs ou K-NNs, escalonar as características (padronização ou normalização) pode acelerar a convergência do algoritmo de otimização (descida de gradiente) e melhorar a interpretabilidade dos coeficientes (quando comparados entre si).
    *   **Tratamento de Outliers**: Outliers podem distorcer os coeficientes. Considere estratégias de tratamento, como remoção ou transformação.
    *   **Variáveis Categóricas**: Use *One-Hot Encoding* para variáveis nominais e *Ordinal Encoding* (com cautela) para variáveis ordinais.
3.  **Seleção de Características**: Embora robusta, a Regressão Logística se beneficia da seleção de características para reduzir o ruído e evitar o overfitting. Técnicas como RFE (Recursive Feature Elimination), Lasso (L1 Regularization) ou análise de importância de características (se aplicável, com modelos de árvore para identificar as mais relevantes) são úteis.
4.  **Regularização**: Use regularização (L1 - Lasso ou L2 - Ridge) para evitar overfitting, especialmente em conjuntos de dados com muitas características ou multicolinearidade. A regularização L1 também pode atuar como um método de seleção de características, zerando coeficientes de características menos importantes.
5.  **Análise de Desempenho Além da Acurácia**: Para problemas de classificação desbalanceados, métricas como Precisão, Recall, F1-Score, Curva ROC, AUC (Area Under the Curve) e matriz de confusão são muito mais informativas que a simples acurácia.
6.  **Ponto de Corte (Threshold)**: A saída da Regressão Logística é uma probabilidade. Um ponto de corte padrão de 0.5 classifica acima como 1 e abaixo como 0. No entanto, você pode ajustar este ponto de corte para otimizar métricas específicas (e.g., recall para detecção de fraudes, precisão para marketing direcionado) com base nos custos e benefícios de falsos positivos e falsos negativos.
7.  **Linearidade dos Log-Odds**: A principal suposição é que existe uma relação linear entre as variáveis independentes e o *logaritmo das odds*. Se essa suposição for violada, transformações de características (e.g., polinomial) ou a inclusão de termos de interação podem ser necessárias.
8.  **Multicolinearidade**: Alta correlação entre variáveis independentes (multicolinearidade) pode tornar os coeficientes instáveis e difíceis de interpretar. Verifique o VIF (Variance Inflation Factor) e considere remover uma das variáveis correlacionadas ou combiná-las.
9.  **Interpretação Causal vs. Correlação**: Lembre-se que a Regressão Logística (e a maioria dos modelos preditivos) estabelece correlações, não causalidade direta. Tenha cuidado ao inferir relações causais sem um design experimental adequado.

## 5. Exemplo Prático: Detecção de Fraude de Cartão de Crédito (Simulado)

Vamos criar um conjunto de dados sintético para ilustrar a Regressão Logística na detecção de fraude.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Gerar dados sintéticos
np.random.seed(42)
num_samples = 1000

# Características:
# transaction_amount: valor da transação
# transaction_hour: hora do dia da transação (0-23)
# user_reputation_score: pontuação de reputação do usuário (maior = melhor)
x1 = np.random.normal(loc=100, scale=50, size=num_samples) # transaction_amount
x2 = np.random.randint(0, 24, size=num_samples) # transaction_hour
x3 = np.random.normal(loc=0.7, scale=0.15, size=num_samples) # user_reputation_score

x3 = np.clip(x3, 0.01, 0.99) # Garante que a reputação esteja entre 0 e 1

# Gerar variável dependente (fraude: 1 = fraude, 0 = não fraude)
# Fraude é mais provável com altos valores, horas incomuns (tarde da noite/madrugada) e baixa reputação

# Log-odds para fraude (AJUSTADO PARA GERAR MAIS FRAUDES)
log_odds_fraud = (
    0.02 * x1  # Aumenta a chance de fraude com valor alto
    + 0.1 * (x2 < 6).astype(int)  # Aumenta a chance de fraude de 0h a 5h
    + 0.1 * (x2 > 22).astype(int) # Aumenta a chance de fraude de 23h a 24h
    - 2 * x3   # Diminui a chance de fraude com boa reputação (menor impacto negativo)
    - 2        # Intercepto ajustado para aumentar a proporção geral de fraude
)

# Converter log-odds para probabilidade usando a função sigmoide
probabilities_fraud = 1 / (1 + np.exp(-log_odds_fraud))

# Classificar como fraude (1) ou não fraude (0) com base na probabilidade
# Adicionar um pouco de ruído para tornar a separação menos perfeita
y = (probabilities_fraud > np.random.rand(num_samples) * 0.9 + 0.05).astype(int)

# Criar DataFrame
df = pd.DataFrame({
    'transaction_amount': x1,
    'transaction_hour': x2,
    'user_reputation_score': x3,
    'is_fraud': y
})

print(f"Número total de transações: {len(df)}")
print(f"Número de transações fraudulentas: {df['is_fraud'].sum()} ({df['is_fraud'].mean():.2%})")
display(df.head())
```

### 5.1. Visualização dos Dados

```python
# Gráfico de dispersão entre 'transaction_amount' e 'user_reputation_score'
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='transaction_amount',
    y='user_reputation_score',
    hue='is_fraud',
    palette='coolwarm',
    alpha=0.6,
    data=df
)
plt.title('Dispersão de Transações por Valor, Reputação e Fraude')
plt.xlabel('Valor da Transação')
plt.ylabel('Pontuação de Reputação do Usuário')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Gráfico de barras para 'transaction_hour'
plt.figure(figsize=(10, 6))
sns.countplot(x='transaction_hour', hue='is_fraud', data=df, palette='viridis')
plt.title('Número de Transações por Hora e Status de Fraude')
plt.xlabel('Hora da Transação')
plt.ylabel('Contagem de Transações')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### 5.2. Treinando o Modelo de Regressão Logística

```python
# Separar características (X) e variável alvo (y)
X = df[['transaction_amount', 'transaction_hour', 'user_reputation_score']]
y = df['is_fraud']

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Inicializar e treinar o modelo de Regressão Logística
model = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' é bom para pequenos datasets e L1/L2
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1] # Probabilidade da classe positiva (fraude)

print("Modelo de Regressão Logística Treinado!")
```

### 5.3. Avaliação do Modelo e Interpretação

```python
print(f"Acurácia: {accuracy_score(y_test, y_pred):.4f}")
print("\nMatriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))

# Coeficientes do modelo
print("\nCoeficientes do Modelo (log-odds):")
for feature, coef in zip(X.columns, model.coef_[0]):
    print(f"  {feature}: {coef:.4f}")
print(f"  Intercepto: {model.intercept_[0]:.4f}")

# Odds Ratios
print("\nOdds Ratios (e^coef):")
for feature, coef in zip(X.columns, model.coef_[0]):
    print(f"  {feature}: {np.exp(coef):.4f}")
```

### 5.4. Curva ROC e AUC

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Classificador Aleatório')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva Característica de Operação do Receptor (ROC)')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
