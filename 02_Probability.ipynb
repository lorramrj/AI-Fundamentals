# Probability Theory

Probability theory is a branch of mathematics concerned with probability. Although there are several different probability interpretations, probability theory deals with the rigorous formalization of probability as a mathematical concept in terms of a set of axioms or definitions.

## Sample Space

The **sample space** (often denoted by $S$ or $\Omega$) of an experiment or random trial is the set of all possible outcomes. For example:

*   When flipping a coin once, the sample space is $S = \{Heads, Tails\}$.
*   When rolling a standard six-sided die, the sample space is $S = \{1, 2, 3, 4, 5, 6\}$.
*   When flipping two coins, the sample space is $S = \{HH, HT, TH, TT\}$.

## Events

An **event** is any subset of the sample space. It's a collection of outcomes from a random experiment. For example, when rolling a six-sided die:

*   The event "rolling an even number" is $E_1 = \{2, 4, 6\}$.
*   The event "rolling a number greater than 4" is $E_2 = \{5, 6\}$.

The probability of an event $E$ occurring is often denoted as $P(E)$. If all outcomes in the sample space are equally likely, then:

$P(E) = \frac{\text{Number of outcomes in E}}{\text{Total number of outcomes in S}}$

## Probability for Equiprobable Finite Events

For finite equiprobable events, the probability of an event $E$ is calculated as:

$P(E) = \frac{\text{Number of favorable outcomes for E}}{\text{Total number of possible outcomes}}$

The `math.comb` and `math.perm` functions help in calculating the 'Number of outcomes' by determining combinations or permutations, depending on whether the order matters.

```python
import math

# Example: Probability of drawing 2 aces from a deck of 52 cards (without replacement)

total_cards = 52
total_aces = 4

# Total number of ways to choose 2 cards from 52 (order doesn't matter)
total_outcomes = math.comb(total_cards, 2)
print(f"Total ways to choose 2 cards from 52: {total_outcomes}")

# Number of ways to choose 2 aces from 4 aces
favorable_outcomes = math.comb(total_aces, 2)
print(f"Ways to choose 2 aces from 4: {favorable_outcomes}")

# Calculate the probability
probability_two_aces = favorable_outcomes / total_outcomes
print(f"Probability of drawing 2 aces: {probability_two_aces:.6f}")

# Example: Probability of picking 2 specific people (A and B) out of 5 people for a specific ordered task

total_people = 5

# Total number of ways to arrange 2 people from 5 (order matters)
total_ordered_outcomes = math.perm(total_people, 2)
print(f"\nTotal ordered arrangements of 2 people from 5: {total_ordered_outcomes}")

# Number of ways to arrange the 2 specific people (A and B) in their specific positions
favorable_ordered_outcomes = math.perm(2, 2) # A then B, or B then A if positions are distinct
print(f"Ways to arrange 2 specific people for 2 specific positions: {favorable_ordered_outcomes}")

probability_specific_arrangement = favorable_ordered_outcomes / total_ordered_outcomes
print(f"Probability of a specific ordered arrangement of 2 people: {probability_specific_arrangement:.6f}")
```

## Independent Events

Two events $A$ and $B$ are **independent** if the occurrence of one does not affect the probability of the other. Mathematically, this means:

$P(A \text{ and } B) = P(A) \cdot P(B)$

or, if $P(B) \neq 0$, $P(A|B) = P(A)$.

**Example**: Flipping a coin twice. The outcome of the first flip does not affect the outcome of the second flip. $P(Heads \text{ on first flip and Heads on second flip}) = P(Heads) \cdot P(Heads) = 0.5 \cdot 0.5 = 0.25$.

## Conditional Probability

**Conditional probability** is the probability of an event occurring given that another event has already occurred. It is denoted as $P(A|B)$, which is the probability of event $A$ given that event $B$ has occurred.

The formula for conditional probability is:

$P(A|B) = \frac{P(A \text{ and } B)}{P(B)}$

provided that $P(B) > 0$.

**Example**: What is the probability of rolling a 6 given that an even number was rolled? ($A$=rolling a 6, $B$=rolling an even number).
$S = \{1, 2, 3, 4, 5, 6\}$
$A = \{6\}$, $P(A) = 1/6$
$B = \{2, 4, 6\}$, $P(B) = 3/6 = 1/2$
$A \text{ and } B = \{6\}$, $P(A \text{ and } B) = 1/6$
$P(A|B) = \frac{1/6}{1/2} = 1/3$.

## Multiplication Theorem

The **multiplication theorem** (or general multiplication rule) provides a way to calculate the probability of two events both occurring. It is derived directly from the conditional probability formula:

$P(A \text{ and } B) = P(A|B) \cdot P(B)$

or equivalently, $P(A \text{ and } B) = P(B|A) \cdot P(A)$.

If events $A$ and $B$ are independent, this simplifies to $P(A \text{ and } B) = P(A) \cdot P(B)$, as seen earlier.

## Total Probability Theorem

The **theorem of total probability** states that if $B_1, B_2, \dots, B_n$ are a partition of the sample space (i.e., they are mutually exclusive events and their union is the entire sample space, $S = B_1 \cup B_2 \cup \dots \cup B_n$), then the probability of any event $A$ can be calculated as:

$P(A) = P(A|B_1)P(B_1) + P(A|B_2)P(B_2) + \dots + P(A|B_n)P(B_n)$

## Useful Python Libraries and Functions

Python offers several libraries that are incredibly useful for working with probability and statistics, for both theoretical calculations and simulations.

```python
import random
import math
from collections import Counter
from scipy import stats

print("### 1. `random` module for simulations ###")

# Simulating a coin flip
coin_flips = [random.choice(['Heads', 'Tails']) for _ in range(1000)]
print(f"Coin flip counts: {Counter(coin_flips)}")

# Simulating a die roll
die_rolls = [random.randint(1, 6) for _ in range(1000)]
print(f"Die roll counts: {Counter(die_rolls)}")

print("\n### 2. `math` module for combinations and permutations ###")

# Combinations (nCr) - number of ways to choose k items from n without regard to order
n_combinations = math.comb(5, 2) # How many ways to choose 2 items from 5
print(f"Combinations (C(5,2)): {n_combinations}")

# Permutations (nPr) - number of ways to arrange k items from n where order matters
n_permutations = math.perm(5, 2) # How many ways to arrange 2 items from 5
print(f"Permutations (P(5,2)): {n_permutations}")

print("\n### 3. `scipy.stats` for probability distributions ###")

# Bernoulli distribution (e.g., single coin flip)
p_success = 0.6 # Probability of success
# Probability mass function (PMF): P(X=k)
prob_0 = stats.bernoulli.pmf(0, p_success) # Probability of failure
prob_1 = stats.bernoulli.pmf(1, p_success) # Probability of success
print(f"Bernoulli (p={p_success}) - P(X=0): {prob_0:.4f}, P(X=1): {prob_1:.4f}")

# Binomial distribution (e.g., number of heads in multiple coin flips)
n_trials = 10
p_heads = 0.5
# P(X=5 heads in 10 flips)
prob_5_heads = stats.binom.pmf(5, n_trials, p_heads)
print(f"Binomial (n={n_trials}, p={p_heads}) - P(X=5 heads): {prob_5_heads:.4f}")

# Normal distribution
mean = 0
std_dev = 1
# Probability density function (PDF): value of the density function at x
density_at_0 = stats.norm.pdf(0, mean, std_dev)
print(f"Normal (mean={mean}, std={std_dev}) - PDF at X=0: {density_at_0:.4f}")

# Cumulative distribution function (CDF): P(X <= x)
prob_less_than_1 = stats.norm.cdf(1, mean, std_dev)
print(f"Normal (mean={mean}, std={std_dev}) - P(X <= 1): {prob_less_than_1:.4f}")

# Inverse CDF (PPF - Percent Point Function): x such that P(X <= x) = q
value_at_95th_percentile = stats.norm.ppf(0.95, mean, std_dev)
print(f"Normal (mean={mean}, std={std_dev}) - 95th percentile value: {value_at_95th_percentile:.4f}")
```

## Random Variables

A **random variable** is a variable whose value is determined by the outcome of a random phenomenon. Random variables can be discrete or continuous.

### Discrete Random Variables

A **discrete random variable** is a random variable that can take on a finite or countably infinite number of values. These values are typically integers or values that can be counted.

**Examples:**
*   The number of heads when flipping a coin 3 times (possible values: 0, 1, 2, 3).
*   The number of defective items in a sample of 10 products (possible values: 0, 1, ..., 10).
*   The number of cars passing a certain point on a road in an hour.

### Common Discrete Probability Distributions

#### 1. Bernoulli Distribution

Represents the probability of success or failure in a single trial. It takes values 1 (success) with probability $p$ and 0 (failure) with probability $1-p$.

**PMF (Probability Mass Function):**
$P(X=k) = p^k (1-p)^{1-k}$ for $k \in \{0, 1\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Bernoulli Distribution Example
p_success = 0.7 # Probability of success
k_values = [0, 1]
pmf_values = [bernoulli.pmf(k, p_success) for k in k_values]

print(f"Bernoulli Distribution (p={p_success}):")
print(f"P(X=0) = {pmf_values[0]:.4f}")
print(f"P(X=1) = {pmf_values[1]:.4f}")

# Plotting Bernoulli PMF
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.bar(k_values, pmf_values, color=['skyblue', 'lightcoral'])
ax.set_xticks(k_values)
ax.set_title(f'Bernoulli PMF (p={p_success})')
ax.set_xlabel('Outcome (k)')
ax.set_ylabel('Probability P(X=k)')
ax.set_ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 2. Binomial Distribution

Represents the number of successes in a fixed number of independent Bernoulli trials. If $n$ is the number of trials and $p$ is the probability of success on each trial.

**PMF (Probability Mass Function):**
$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$ for $k \in \{0, 1, \dots, n\}$

```python
from scipy.stats import binom

# Binomial Distribution Example
n_trials = 10  # Number of trials
p_success_binom = 0.5 # Probability of success in each trial

# Calculate PMF for k from 0 to n_trials
k_values_binom = np.arange(0, n_trials + 1)
pmf_values_binom = [binom.pmf(k, n_trials, p_success_binom) for k in k_values_binom]

print(f"Binomial Distribution (n={n_trials}, p={p_success_binom}):")
for k, pmf in zip(k_values_binom, pmf_values_binom):
    print(f"P(X={k}) = {pmf:.4f}")

# Plotting Binomial PMF
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.bar(k_values_binom, pmf_values_binom, color='lightgreen')
ax.set_title(f'Binomial PMF (n={n_trials}, p={p_success_binom})')
ax.set_xlabel('Number of Successes (k)')
ax.set_ylabel('Probability P(X=k)')
ax.set_xticks(k_values_binom)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 3. Poisson Distribution

Models the number of events occurring in a fixed interval of time or space, given a constant average rate of occurrence ($\lambda$).

**PMF (Probability Mass Function):**
$P(X=k) = \frac{e^{-\lambda} \lambda^k}{k!}$ for $k \in \{0, 1, 2, \dots\}$

```python
from scipy.stats import poisson

# Poisson Distribution Example
lambda_rate = 3 # Average rate of events

# Calculate PMF for k from 0 to 10 (a reasonable range for lambda=3)
k_values_poisson = np.arange(0, 11)
pmf_values_poisson = [poisson.pmf(k, lambda_rate) for k in k_values_poisson]

print(f"Poisson Distribution (lambda={lambda_rate}):")
for k, pmf in zip(k_values_poisson, pmf_values_poisson):
    print(f"P(X={k}) = {pmf:.4f}")

# Plotting Poisson PMF
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.bar(k_values_poisson, pmf_values_poisson, color='salmon')
ax.set_title(f'Poisson PMF (lambda={lambda_rate})')
ax.set_xlabel('Number of Events (k)')
ax.set_ylabel('Probability P(X=k)')
ax.set_xticks(k_values_poisson)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Continuous Random Variables

A **continuous random variable** is a random variable that can take on any value within a given interval. These are typically measurements.

**Examples:**
*   The height of a randomly selected person.
*   The time it takes for a bus to arrive.
*   The temperature in a city.

### Common Continuous Probability Distributions

For continuous distributions, we talk about **PDF (Probability Density Function)** instead of PMF. The probability of a continuous random variable taking on any *single* specific value is 0. Instead, we calculate probabilities over intervals, which correspond to the area under the PDF curve.

#### 1. Uniform Distribution

Represents events that have an equal probability of occurring over a given interval $[a, b]$.

**PDF (Probability Density Function):**
$f(x) = \frac{1}{b-a}$ for $a \le x \le b$, and $0$ otherwise.

```python
from scipy.stats import uniform

# Uniform Distribution Example
a = 0  # Lower bound
b = 10 # Upper bound

x_values_uniform = np.linspace(a - 1, b + 1, 500)
pdf_values_uniform = uniform.pdf(x_values_uniform, loc=a, scale=b-a)

print(f"Uniform Distribution (a={a}, b={b}):")
print(f"PDF is 1/({b}-{a}) = {1/(b-a):.2f} for x in [{a}, {b}], and 0 otherwise.")

# Plotting Uniform PDF
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x_values_uniform, pdf_values_uniform, color='purple', linewidth=2)
ax.fill_between(x_values_uniform, 0, pdf_values_uniform, color='purple', alpha=0.2)
ax.set_title(f'Uniform PDF (a={a}, b={b})')
ax.set_xlabel('Value (x)')
ax.set_ylabel('Probability Density f(x)')
ax.set_ylim(bottom=0)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 2. Normal (Gaussian) Distribution

One of the most important distributions, characterized by its symmetric, bell-shaped curve. It's defined by its mean ($\mu$) and standard deviation ($\sigma$).

**PDF (Probability Density Function):**
$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left(\frac{x-\mu}{\sigma}\right)^2}$

```python
from scipy.stats import norm

# Normal Distribution Example
mean_norm = 0
std_dev_norm = 1 # Standard Normal Distribution

x_values_norm = np.linspace(-4, 4, 500)
pdf_values_norm = norm.pdf(x_values_norm, loc=mean_norm, scale=std_dev_norm)

print(f"Normal Distribution (mu={mean_norm}, sigma={std_dev_norm}):")
print(f"PDF at X=0: {norm.pdf(0, loc=mean_norm, scale=std_dev_norm):.4f}")
print(f"P(X <= 1): {norm.cdf(1, loc=mean_norm, scale=std_dev_norm):.4f}") # CDF

# Plotting Normal PDF
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x_values_norm, pdf_values_norm, color='darkblue', linewidth=2)
ax.fill_between(x_values_norm, 0, pdf_values_norm, color='darkblue', alpha=0.1)
ax.set_title(f'Normal PDF (mu={mean_norm}, sigma={std_dev_norm})')
ax.set_xlabel('Value (x)')
ax.set_ylabel('Probability Density f(x)')
ax.set_ylim(bottom=0)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 3. Exponential Distribution

Models the time until an event occurs in a Poisson process (where events occur continuously and independently at a constant average rate). It is often used for waiting times.

**PDF (Probability Density Function):**
$f(x) = \lambda e^{-\lambda x}$ for $x \ge 0$, and $0$ otherwise. ($\lambda$ is the rate parameter, the inverse of the mean waiting time).

```python
from scipy.stats import expon

# Exponential Distribution Example
lambda_exp = 0.5 # Rate parameter (e.g., events per unit time)

# In scipy.stats, scale = 1/lambda
scale_exp = 1 / lambda_exp

x_values_exp = np.linspace(0, 10, 500)
pdf_values_exp = expon.pdf(x_values_exp, scale=scale_exp)

print(f"Exponential Distribution (lambda={lambda_exp}):")
print(f"PDF at X=1: {expon.pdf(1, scale=scale_exp):.4f}")
print(f"P(X <= 2): {expon.cdf(2, scale=scale_exp):.4f}") # CDF

# Plotting Exponential PDF
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
ax.plot(x_values_exp, pdf_values_exp, color='darkgreen', linewidth=2)
ax.fill_between(x_values_exp, 0, pdf_values_exp, color='darkgreen', alpha=0.1)
ax.set_title(f'Exponential PDF (lambda={lambda_exp})')
ax.set_xlabel('Value (x)')
ax.set_ylabel('Probability Density f(x)')
ax.set_ylim(bottom=0)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Expected Value, Variance, and Standard Deviation

These are key measures used to describe the central tendency and spread of a probability distribution.

### Expected Value (Mean)

The **Expected Value** (or mean, $E[X]$ or $\mu$) of a random variable $X$ is the weighted average of all possible values that the random variable can take. It represents the long-run average value of the random variable if the experiment is repeated many times.

*   **For a Discrete Random Variable:**
    $E[X] = \sum_{x} x P(X=x)$

*   **For a Continuous Random Variable:**
    $E[X] = \int_{-\infty}^{\infty} x f(x) dx$

### Variance

The **Variance** ($Var(X)$ or $\sigma^2$) measures the spread or dispersion of the probability distribution. It quantifies how much the values of the random variable typically deviate from the mean.

$Var(X) = E[(X - E[X])^2] = E[X^2] - (E[X])^2$

*   **For a Discrete Random Variable:**
    $Var(X) = \sum_{x} (x - \mu)^2 P(X=x)$

*   **For a Continuous Random Variable:**
    $Var(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) dx$

### Standard Deviation

The **Standard Deviation** ($\sigma$) is the square root of the variance. It is a more interpretable measure of spread than variance because it is in the same units as the random variable itself.

$\sigma = \sqrt{Var(X)}$

### Calculations for Each Distribution

Let's calculate these values for each of the distributions we discussed.

#### 1. Bernoulli Distribution

*   **Expected Value:** $E[X] = p$
*   **Variance:** $Var(X) = p(1-p)$
*   **Standard Deviation:** $\sigma = \sqrt{p(1-p)}$

```python
from scipy.stats import bernoulli

p_success = 0.7 # from previous example

# Using formulas
bern_mean_formula = p_success
bern_var_formula = p_success * (1 - p_success)
bern_std_formula = math.sqrt(bern_var_formula)

# Using scipy.stats methods
bern_mean_scipy = bernoulli.mean(p_success)
bern_var_scipy = bernoulli.var(p_success)
bern_std_scipy = bernoulli.std(p_success)

print(f"Bernoulli Distribution (p={p_success}):")
print(f"  Expected Value (Formula): {bern_mean_formula:.4f}, (Scipy): {bern_mean_scipy:.4f}")
print(f"  Variance (Formula): {bern_var_formula:.4f}, (Scipy): {bern_var_scipy:.4f}")
print(f"  Standard Deviation (Formula): {bern_std_formula:.4f}, (Scipy): {bern_std_scipy:.4f}")
```

#### 2. Binomial Distribution

*   **Expected Value:** $E[X] = np$
*   **Variance:** $Var(X) = np(1-p)$
*   **Standard Deviation:** $\sigma = \sqrt{np(1-p)}$

```python
from scipy.stats import binom

n_trials = 10     # from previous example
p_success_binom = 0.5 # from previous example

# Using formulas
binom_mean_formula = n_trials * p_success_binom
binom_var_formula = n_trials * p_success_binom * (1 - p_success_binom)
binom_std_formula = math.sqrt(binom_var_formula)

# Using scipy.stats methods
binom_mean_scipy = binom.mean(n_trials, p_success_binom)
binom_var_scipy = binom.var(n_trials, p_success_binom)
binom_std_scipy = binom.std(n_trials, p_success_binom)

print(f"Binomial Distribution (n={n_trials}, p={p_success_binom}):")
print(f"  Expected Value (Formula): {binom_mean_formula:.4f}, (Scipy): {binom_mean_scipy:.4f}")
print(f"  Variance (Formula): {binom_var_formula:.4f}, (Scipy): {binom_var_scipy:.4f}")
print(f"  Standard Deviation (Formula): {binom_std_formula:.4f}, (Scipy): {binom_std_scipy:.4f}")
```

#### 3. Poisson Distribution

*   **Expected Value:** $E[X] = \lambda$
*   **Variance:** $Var(X) = \lambda$
*   **Standard Deviation:** $\sigma = \sqrt{\lambda}$

```python
from scipy.stats import poisson

lambda_rate = 3 # from previous example

# Using formulas
poisson_mean_formula = lambda_rate
poisson_var_formula = lambda_rate
poisson_std_formula = math.sqrt(poisson_var_formula)

# Using scipy.stats methods
poisson_mean_scipy = poisson.mean(lambda_rate)
poisson_var_scipy = poisson.var(lambda_rate)
poisson_std_scipy = poisson.std(lambda_rate)

print(f"Poisson Distribution (lambda={lambda_rate}):")
print(f"  Expected Value (Formula): {poisson_mean_formula:.4f}, (Scipy): {poisson_mean_scipy:.4f}")
print(f"  Variance (Formula): {poisson_var_formula:.4f}, (Scipy): {poisson_var_scipy:.4f}")
print(f"  Standard Deviation (Formula): {poisson_std_formula:.4f}, (Scipy): {poisson_std_scipy:.4f}")
```

#### 4. Uniform Distribution

For a uniform distribution over the interval $[a, b]$:

*   **Expected Value:** $E[X] = \frac{a+b}{2}$
*   **Variance:** $Var(X) = \frac{(b-a)^2}{12}$
*   **Standard Deviation:** $\sigma = \sqrt{\frac{(b-a)^2}{12}}$

```python
from scipy.stats import uniform

a = 0   # from previous example
b = 10  # from previous example

# Using formulas
uniform_mean_formula = (a + b) / 2
uniform_var_formula = ((b - a)**2) / 12
uniform_std_formula = math.sqrt(uniform_var_formula)

# Using scipy.stats methods (loc=a, scale=b-a)
uniform_mean_scipy = uniform.mean(loc=a, scale=b-a)
uniform_var_scipy = uniform.var(loc=a, scale=b-a)
uniform_std_scipy = uniform.std(loc=a, scale=b-a)

print(f"Uniform Distribution (a={a}, b={b}):")
print(f"  Expected Value (Formula): {uniform_mean_formula:.4f}, (Scipy): {uniform_mean_scipy:.4f}")
print(f"  Variance (Formula): {uniform_var_formula:.4f}, (Scipy): {uniform_var_scipy:.4f}")
print(f"  Standard Deviation (Formula): {uniform_std_formula:.4f}, (Scipy): {uniform_std_scipy:.4f}")
```

#### 5. Normal (Gaussian) Distribution

For a normal distribution with mean $\mu$ and standard deviation $\sigma$:

*   **Expected Value:** $E[X] = \mu$
*   **Variance:** $Var(X) = \sigma^2$
*   **Standard Deviation:** $\sigma = \sigma$

```python
from scipy.stats import norm

mean_norm = 0      # from previous example
std_dev_norm = 1   # from previous example

# Using formulas
norm_mean_formula = mean_norm
norm_var_formula = std_dev_norm**2
norm_std_formula = std_dev_norm

# Using scipy.stats methods (loc=mean, scale=std_dev)
norm_mean_scipy = norm.mean(loc=mean_norm, scale=std_dev_norm)
norm_var_scipy = norm.var(loc=mean_norm, scale=std_dev_norm)
norm_std_scipy = norm.std(loc=mean_norm, scale=std_dev_norm)

print(f"Normal Distribution (mu={mean_norm}, sigma={std_dev_norm}):")
print(f"  Expected Value (Formula): {norm_mean_formula:.4f}, (Scipy): {norm_mean_scipy:.4f}")
print(f"  Variance (Formula): {norm_var_formula:.4f}, (Scipy): {norm_var_scipy:.4f}")
print(f"  Standard Deviation (Formula): {norm_std_formula:.4f}, (Scipy): {norm_std_scipy:.4f}")
```

#### 6. Exponential Distribution

For an exponential distribution with rate parameter $\lambda$:

*   **Expected Value:** $E[X] = \frac{1}{\lambda}$
*   **Variance:** $Var(X) = \frac{1}{\lambda^2}$
*   **Standard Deviation:** $\sigma = \frac{1}{\lambda}$

```python
from scipy.stats import expon

lambda_exp = 0.5 # from previous example

# Using formulas
exp_mean_formula = 1 / lambda_exp
exp_var_formula = 1 / (lambda_exp**2)
exp_std_formula = 1 / lambda_exp

# Using scipy.stats methods (scale=1/lambda)
# Note: scipy.stats uses 'scale' parameter which is 1/lambda
scale_exp_scipy = 1 / lambda_exp
exp_mean_scipy = expon.mean(scale=scale_exp_scipy)
exp_var_scipy = expon.var(scale=scale_exp_scipy)
exp_std_scipy = expon.std(scale=scale_exp_scipy)

print(f"Exponential Distribution (lambda={lambda_exp}):")
print(f"  Expected Value (Formula): {exp_mean_formula:.4f}, (Scipy): {exp_mean_scipy:.4f}")
print(f"  Variance (Formula): {exp_var_formula:.4f}, (Scipy): {exp_var_scipy:.4f}")
print(f"  Standard Deviation (Formula): {exp_std_formula:.4f}, (Scipy): {exp_std_scipy:.4f}")
```
