### Funções do Módulo `math` para Combinatória

O módulo `math` do Python oferece funções diretas para calcular fatoriais, permutações e combinações, que são a base da combinatória.

*   **`math.factorial(n)`**: Retorna o fatorial de `n`. O fatorial de `n` é o produto de todos os inteiros positivos menores ou iguais a `n` ($n! = n \times (n-1) \times \ldots \times 1$).
    *   **Exemplo:**
    ```python
    import math
    print(f"Fatorial de 5: {math.factorial(5)}") # Saída: Fatorial de 5: 120
    ```

*   **`math.perm(n, k=None)`**: Retorna o número de maneiras de escolher `k` itens de `n` itens sem repetição e com ordem. Se `k` for `None` (ou omitido), ele retorna o número total de permutações de `n` itens (ou seja, `n!`). É equivalente a $\frac{n!}{(n-k)!}$.
    *   **Exemplo:**
    ```python
    import math
    print(f"Permutações de 5 itens tomados 2 a 2: {math.perm(5, 2)}") # Saída: Permutações de 5 itens tomados 2 a 2: 20
    print(f"Permutações de 3 itens (todas): {math.perm(3)}") # Saída: Permutações de 3 itens (todas): 6
    ```

*   **`math.comb(n, k)`**: Retorna o número de maneiras de escolher `k` itens de `n` itens sem repetição e sem ordem. É equivalente a $\frac{n!}{k!(n-k)!}$.
    *   **Exemplo:**
    ```python
    import math
    print(f"Combinações de 5 itens tomados 2 a 2: {math.comb(5, 2)}") # Saída: Combinações de 5 itens tomados 2 a 2: 10
    ```

Essas funções são muito úteis para cálculos rápidos e precisos em problemas de combinatória.

### Módulo `itertools`

O módulo `itertools` é uma parte da biblioteca padrão do Python que fornece ferramentas para criar iteradores de forma eficiente para loops. Ele é especialmente útil para trabalhar com sequências e para tarefas de combinatória, como as que exploramos.

Algumas das funções mais relevantes para combinatória incluem:

*   **`itertools.permutations(iterable, r=None)`**: Retorna um iterador que produz permutações de elementos de `iterable` de comprimento `r`. Se `r` não for especificado ou for `None`, o comprimento de `iterable` é usado.
    *   **Exemplo:** `list(itertools.permutations(['A', 'B', 'C'], 2))` -> `[('A', 'B'), ('A', 'C'), ('B', 'A'), ('B', 'C'), ('C', 'A'), ('C', 'B')]`

*   **`itertools.combinations(iterable, r)`**: Retorna um iterador que produz combinações de elementos de `iterable` de comprimento `r`. A ordem não importa e não há elementos repetidos.
    *   **Exemplo:** `list(itertools.combinations(['A', 'B', 'C', 'D'], 2))` -> `[('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'C'), ('B', 'D'), ('C', 'D')]`

*   **`itertools.product(*iterables, repeat=1)`**: Retorna um iterador que produz o produto cartesiano dos `iterables` de entrada. É equivalente a aninhar vários loops `for`. É útil para arranjos com repetição.
    *   **Exemplo:** `list(itertools.product(['A', 'B'], repeat=2))` -> `[('A', 'A'), ('A', 'B'), ('B', 'A'), ('B', 'B')]`

*   **`itertools.combinations_with_replacement(iterable, r)`**: Retorna um iterador que produz combinações de elementos de `iterable` de comprimento `r`, onde a repetição é permitida.
    *   **Exemplo:** `list(itertools.combinations_with_replacement(['A', 'B'], 2))` -> `[('A', 'A'), ('A', 'B'), ('B', 'B')]`

Além dessas, `itertools` possui muitas outras funções para criar sequências infinitas, agrupamentos e outras operações iteráveis, como `count()`, `cycle()`, `repeat()`, `chain()`, `groupby()`, `zip_longest()`, entre outras. É uma ferramenta fundamental para programação funcional e para lidar com grandes conjuntos de dados de forma eficiente em Python.

### Permutação Simples

A permutação simples refere-se ao número de maneiras de organizar $n$ objetos distintos em uma sequência, onde a ordem importa e cada objeto é usado apenas uma vez. A fórmula para permutar $n$ objetos tomados $k$ a $k$ é $P(n, k) = \frac{n!}{(n-k)!}$. Se $k=n$, é $n!$.

````python
import math

# Exemplo: Permutação Simples de 3 itens (A, B, C)
items = ['A', 'B', 'C']

# Permutação de todos os itens (3!)
n_simple_perm_all = math.perm(len(items))
print(f"Permutação simples de todos os {len(items)} itens: = {n_simple_perm_all}")

# Permutação de 3 itens tomados 2 a 2 (P(3, 2))
n_simple_perm_k = math.perm(len(items), 2)
print(f"Permutação simples de {len(items)} itens tomados 2 a 2: {n_simple_perm_k}")

# Usando itertools para listar as permutações
import itertools

list_items_simple = ['A', 'B', 'C']
permutations_simple = list(itertools.permutations(list_items_simple))
print(f"Permutações simples de {list_items_simple}: {permutations_simple}")

permutations_simple_k = list(itertools.permutations(list_items_simple, 2))
print(f"Permutações simples de {list_items_simple} tomadas 2 a 2: {permutations_simple_k}")
````

### Permutação com Repetição

A permutação com repetição refere-se ao número de maneiras de organizar $n$ objetos, onde alguns são idênticos. Se você tem $n$ objetos com $n_1$ de um tipo, $n_2$ de outro tipo, ..., $n_k$ de um $k$-ésimo tipo, o número de permutações é $\frac{n!}{n_1! n_2! \ldots n_k!}$.

Se a intenção for arranjos com repetição onde você pode escolher $k$ itens de $n$ tipos com repetição, a fórmula é $n^k$ (trataremos isso como 'Arranjo com Repetição'). Para gerar as sequências, podemos usar `itertools.product`.

````python
import math
import itertools
from collections import Counter

# Exemplo: Permutação de letras na palavra "MATEMATICA"
word = "MATEMATICA"
char_counts = Counter(word)

n = len(word)

denominator = 1
for count in char_counts.values():
    denominator *= math.factorial(count)

permutations_with_repetition_count = math.factorial(n) / denominator
print(f"Número de permutações com repetição para '{word}': {int(permutations_with_repetition_count)}")

# Exemplo para gerar permutações com repetição (escolhendo k de n, com reposição)
# Isso é mais alinhado com 'Arranjo com Repetição' na maioria das interpretações
items_for_repetition = ['A', 'B']
k_repetition = 3

# Gerar sequências de tamanho k com elementos dos items_for_repetition, com repetição
permutations_with_replacement = list(itertools.product(items_for_repetition, repeat=k_repetition))
print(f"Permutações com repetição (arranjos com repetição) de {items_for_repetition} tomados {k_repetition} a {k_repetition}: {permutations_with_replacement}")
````

### Permutação Circular

A permutação circular refere-se ao número de maneiras de organizar $n$ objetos distintos em um círculo. A fórmula é $P_c(n) = (n-1)!$.

````python
import math

# Exemplo: Permutação Circular de 4 pessoas
n_circular = 4

if n_circular > 0:
    circular_permutations = math.factorial(n_circular - 1)
    print(f"Permutação circular de {n_circular} itens: {circular_permutations}")
else:
    print("Não é possível calcular permutação circular para n <= 0.")
````

### Arranjo Simples

O arranjo simples é equivalente à permutação simples de $n$ objetos tomados $k$ a $k$. A ordem importa e não há repetição. A fórmula é $A(n, k) = \frac{n!}{(n-k)!}$, que é o mesmo que $P(n, k)$.

````python
import math
import itertools

# Exemplo: Arranjo Simples de 5 itens tomados 2 a 2
n_arr = 5
k_arr = 2

arrangement_count = math.perm(n_arr, k_arr)
print(f"Arranjo simples de {n_arr} itens tomados {k_arr} a {k_arr}: {arrangement_count}")

# Usando itertools para listar os arranjos
list_items_arr = [1, 2, 3, 4, 5]
arrangements_simple = list(itertools.permutations(list_items_arr, k_arr))
print(f"Arranjos simples de {list_items_arr} tomados {k_arr} a {k_arr}: {arrangements_simple}")
````

### Arranjo com Repetição

O arranjo com repetição refere-se ao número de maneiras de escolher $k$ objetos de um conjunto de $n$ objetos distintos, onde a ordem importa e a repetição é permitida. A fórmula é $A_r(n, k) = n^k$.

````python
import itertools

# Exemplo: Arranjo com Repetição de 3 dígitos (0, 1, 2) tomados 2 a 2 (senhas de 2 dígitos)
n_rep = 3 # número de opções
k_rep = 2 # tamanho do arranjo

arrangement_with_repetition_count = n_rep ** k_rep
print(f"Arranjo com repetição de {n_rep} itens tomados {k_rep} a {k_rep}: {arrangement_with_repetition_count}")

# Usando itertools para listar os arranjos com repetição
items_for_arr_rep = [0, 1, 2]
arrangements_with_repetition = list(itertools.product(items_for_arr_rep, repeat=k_rep))
print(f"Arranjos com repetição de {items_for_arr_rep} tomados {k_rep} a {k_rep}: {arrangements_with_repetition}")
````

### Combinação Simples

A combinação simples refere-se ao número de maneiras de escolher $k$ objetos de um conjunto de $n$ objetos distintos, onde a ordem não importa e a repetição não é permitida. A fórmula é $C(n, k) = \frac{n!}{k!(n-k)!}$.

````python
import math
import itertools

# Exemplo: Combinação Simples de 5 itens tomados 2 a 2
n_comb = 5
k_comb = 2

combinations_count = math.comb(n_comb, k_comb)
print(f"Combinação simples de {n_comb} itens tomados {k_comb} a {k_comb}: {combinations_count}")

# Usando itertools para listar as combinações
list_items_comb = ['A', 'B', 'C', 'D', 'E']
combinations_simple = list(itertools.combinations(list_items_comb, k_comb))
print(f"Combinações simples de {list_items_comb} tomadas {k_comb} a {k_comb}: {combinations_simple}")
````

### Combinação com Repetição

A combinação com repetição refere-se ao número de maneiras de escolher $k$ objetos de um conjunto de $n$ objetos distintos, onde a ordem não importa e a repetição é permitida. A fórmula é $C_r(n, k) = C(n+k-1, k) = \frac{(n+k-1)!}{k!(n-1)!}$.

````python
import math
import itertools

# Exemplo: Combinação com Repetição de 3 tipos de frutas (maçã, banana, laranja)
# escolhendo 2 frutas (m, b, l)
# n = 3 (tipos de frutas), k = 2 (frutas a escolher)
n_comb_rep = 3
k_comb_rep = 2

# Usando a fórmula C(n+k-1, k)
combinations_with_repetition_count = math.comb(n_comb_rep + k_comb_rep - 1, k_comb_rep)
print(f"Combinação com repetição de {n_comb_rep} tipos de itens tomados {k_comb_rep} a {k_comb_rep}: {combinations_with_repetition_count}")

# Usando itertools para listar as combinações com repetição
items_for_comb_rep = ['maçã', 'banana', 'laranja']
combinations_with_repetition = list(itertools.combinations_with_replacement(items_for_comb_rep, k_comb_rep))
print(f"Combinações com repetição de {items_for_comb_rep} tomadas {k_comb_rep} a {k_comb_rep}: {combinations_with_repetition}")
````

### Outras Funções Úteis do `itertools`

Além das funções de combinatória, o `itertools` oferece outras ferramentas poderosas para trabalhar com iteradores de forma eficiente.

*   **`itertools.count(start=0, step=1)`**: Retorna um iterador que produz uma sequência de números inteiros, começando em `start` e incrementando por `step`. É uma sequência infinita.
    *   **Exemplo:**
    ```python
    import itertools
    # Gerar os primeiros 5 números a partir de 10 com passo 2
    for i in itertools.count(10, 2):
        if i > 20: break
        print(i)
    ```

*   **`itertools.cycle(iterable)`**: Retorna um iterador que repete os elementos de um `iterable` infinitamente.
    *   **Exemplo:**
    ```python
    import itertools
    count = 0
    for item in itertools.cycle(['A', 'B', 'C']):
        if count > 5: break
        print(item)
        count += 1
    ```

*   **`itertools.repeat(object, times=None)`**: Retorna um iterador que repete um `object` um determinado número de `times`, ou infinitamente se `times` não for especificado.
    *   **Exemplo:**
    ```python
    import itertools
    for i in itertools.repeat('Olá', 3):
        print(i)
    ```

*   **`itertools.chain(*iterables)`**: Retorna um iterador que encadeia vários iteráveis em uma única sequência.
    *   **Exemplo:**
    ```python
    import itertools
    list1 = [1, 2, 3]
    list2 = ['a', 'b']
    chained = list(itertools.chain(list1, list2))
    print(chained)
    ```

*   **`itertools.groupby(iterable, key=None)`**: Retorna um iterador que agrupa elementos consecutivos do `iterable` que têm a mesma chave (definida pela função `key`). O `iterable` deve estar **ordenado** pela chave para que `groupby` funcione corretamente.
    *   **Exemplo:**
    ```python
    import itertools

    data = [('A', 1), ('A', 2), ('B', 3), ('B', 4), ('A', 5)]
    # Para groupby funcionar corretamente, os dados devem ser pré-ordenados pela chave
    data_sorted = sorted(data, key=lambda x: x[0])

    for key, group in itertools.groupby(data_sorted, lambda x: x[0]):
        print(f"Key: {key}, Group: {list(group)}")
    ```
