# Combinatorics

### `math` Module Functions for Combinatorics

Python's `math` module provides direct functions for calculating factorials, permutations, and combinations, which are the foundation of combinatorics.

*   **`math.factorial(n)`**: Returns the factorial of `n`. The factorial of `n` is the product of all positive integers less than or equal to `n` ($n! = n \times (n-1) \times \ldots \times 1$).
    *   **Example:**
    ```python
    import math
    print(f"Factorial of 5: {math.factorial(5)}") # Output: Factorial of 5: 120
    ```

*   **`math.perm(n, k=None)`**: Returns the number of ways to choose `k` items from `n` items without repetition and with order. If `k` is `None` (or omitted), it returns the total number of permutations of `n` items (i.e., `n!`). It is equivalent to $\frac{n!}{(n-k)!}$.
    *   **Example:**
    ```python
    import math
    print(f"Permutations of 5 items taken 2 at a time: {math.perm(5, 2)}") # Output: Permutations of 5 items taken 2 at a time: 20
    print(f"Permutations of 3 items (all): {math.perm(3)}") # Output: Permutations of 3 items (all): 6
    ```

*   **`math.comb(n, k)`**: Returns the number of ways to choose `k` items from `n` items without repetition and without order. It is equivalent to $\frac{n!}{k!(n-k)!}$.
    *   **Example:**
    ```python
    import math
    print(f"Combinations of 5 items taken 2 at a time: {math.comb(5, 2)}") # Output: Combinations of 5 items taken 2 at a time: 10
    ```

These functions are very useful for quick and accurate calculations in combinatorics problems.

### `itertools` Module

The `itertools` module is part of Python's standard library that provides tools to create efficient iterators for loops. It is especially useful for working with sequences and for combinatorics tasks, such as those we have explored.

Some of the most relevant functions for combinatorics include:

*   **`itertools.permutations(iterable, r=None)`**: Returns an iterator that produces permutations of elements from `iterable` of length `r`. If `r` is not specified or is `None`, the length of `iterable` is used.
    *   **Example:** `list(itertools.permutations(['A', 'B', 'C'], 2))` -> `[('A', 'B'), ('A', 'C'), ('B', 'A'), ('B', 'C'), ('C', 'A'), ('C', 'B')]`

*   **`itertools.combinations(iterable, r)`**: Returns an iterator that produces combinations of elements from `iterable` of length `r`. Order does not matter and there are no repeated elements.
    *   **Example:** `list(itertools.combinations(['A', 'B', 'C', 'D'], 2))` -> `[('A', 'B'), ('A', 'C'), ('A', 'D'), ('B', 'C'), ('B', 'D'), ('C', 'D')]`

*   **`itertools.product(*iterables, repeat=1)`**: Returns an iterator that produces the Cartesian product of the input `iterables`. It is equivalent to nesting multiple `for` loops. It is useful for arrangements with repetition.
    *   **Example:** `list(itertools.product(['A', 'B'], repeat=2))` -> `[('A', 'A'), ('A', 'B'), ('B', 'A'), ('B', 'B')]`

*   **`itertools.combinations_with_replacement(iterable, r)`**: Returns an iterator that produces combinations of elements from `iterable` of length `r`, where repetition is allowed.
    *   **Example:** `list(itertools.combinations_with_replacement(['A', 'B'], 2))` -> `[('A', 'A'), ('A', 'B'), ('B', 'B')]`

In addition to these, `itertools` has many other functions for creating infinite sequences, groupings, and other iterable operations, such as `count()`, `cycle()`, `repeat()`, `chain()`, `groupby()`, `zip_longest()`, among others. It is a fundamental tool for functional programming and for efficiently handling large datasets in Python.

### Simple Permutation

Simple permutation refers to the number of ways to arrange $n$ distinct objects in a sequence, where order matters and each object is used only once. The formula to permute $n$ objects taken $k$ at a time is $P(n, k) = \frac{n!}{(n-k)!}$. If $k=n$, it is $n!$.

````python
import math

# Example: Simple Permutation of 3 items (A, B, C)
items = ['A', 'B', 'C']

# Permutation of all items (3!)
n_simple_perm_all = math.perm(len(items))
print(f"Simple permutation of all {len(items)} items: = {n_simple_perm_all}")

# Permutation of 3 items taken 2 at a time (P(3, 2))
n_simple_perm_k = math.perm(len(items), 2)
print(f"Simple permutation of {len(items)} items taken 2 at a time: {n_simple_perm_k}")

# Using itertools to list permutations
import itertools

list_items_simple = ['A', 'B', 'C']
permutations_simple = list(itertools.permutations(list_items_simple))
print(f"Simple permutations of {list_items_simple}: {permutations_simple}")

permutations_simple_k = list(itertools.permutations(list_items_simple, 2))
print(f"Simple permutations of {list_items_simple} taken 2 at a time: {permutations_simple_k}")
````

### Permutation with Repetition

Permutation with repetition refers to the number of ways to arrange $n$ objects where some are identical. If you have $n$ objects with $n_1$ of one type, $n_2$ of another type, ..., $n_k$ of a $k$-th type, the number of permutations is $\frac{n!}{n_1! n_2! \ldots n_k!}$.

If the intention is arrangements with repetition where you can choose $k$ items from $n$ types with repetition, the formula is $n^k$ (we will treat this as 'Arrangement with Repetition'). To generate the sequences, we can use `itertools.product`.

````python
import math
import itertools
from collections import Counter

# Example: Permutation of letters in the word "MATHEMATICS"
word = "MATHEMATICS"
char_counts = Counter(word)

n = len(word)

denominator = 1
for count in char_counts.values():
    denominator *= math.factorial(count)

permutations_with_repetition_count = math.factorial(n) / denominator
print(f"Number of permutations with repetition for '{word}': {int(permutations_with_repetition_count)}")

# Example to generate permutations with repetition (choosing k from n, with replacement)
# This is more aligned with 'Arrangement with Repetition' in most interpretations
items_for_repetition = ['A', 'B']
k_repetition = 3

# Generate sequences of size k with elements from items_for_repetition, with repetition
permutations_with_replacement = list(itertools.product(items_for_repetition, repeat=k_repetition))
print(f"Permutations with repetition (arrangements with repetition) of {items_for_repetition} taken {k_repetition} at a time: {permutations_with_replacement}")
````

### Circular Permutation

Circular permutation refers to the number of ways to arrange $n$ distinct objects in a circle. The formula is $P_c(n) = (n-1)!$.

````python
import math

# Example: Circular Permutation of 4 people
n_circular = 4

if n_circular > 0:
    circular_permutations = math.factorial(n_circular - 1)
    print(f"Circular permutation of {n_circular} items: {circular_permutations}")
else:
    print("Cannot calculate circular permutation for n <= 0.")
````

### Simple Arrangement

Simple arrangement is equivalent to the simple permutation of $n$ objects taken $k$ at a time. Order matters and there is no repetition. The formula is $A(n, k) = \frac{n!}{(n-k)!}$, which is the same as $P(n, k)$.

````python
import math
import itertools

# Example: Simple Arrangement of 5 items taken 2 at a time
n_arr = 5
k_arr = 2

arrangement_count = math.perm(n_arr, k_arr)
print(f"Simple arrangement of {n_arr} items taken {k_arr} at a time: {arrangement_count}")

# Using itertools to list arrangements
list_items_arr = [1, 2, 3, 4, 5]
arrangements_simple = list(itertools.permutations(list_items_arr, k_arr))
print(f"Simple arrangements of {list_items_arr} taken {k_arr} at a time: {arrangements_simple}")
````

### Arrangement with Repetition

Arrangement with repetition refers to the number of ways to choose $k$ objects from a set of $n$ distinct objects, where order matters and repetition is allowed. The formula is $A_r(n, k) = n^k$.

````python
import itertools

# Example: Arrangement with Repetition of 3 digits (0, 1, 2) taken 2 at a time (2-digit passwords)
n_rep = 3 # number of options
k_rep = 2 # size of arrangement

arrangement_with_repetition_count = n_rep ** k_rep
print(f"Arrangement with repetition of {n_rep} items taken {k_rep} at a time: {arrangement_with_repetition_count}")

# Using itertools to list arrangements with repetition
items_for_arr_rep = [0, 1, 2]
arrangements_with_repetition = list(itertools.product(items_for_arr_rep, repeat=k_rep))
print(f"Arrangements with repetition of {items_for_arr_rep} taken {k_rep} at a time: {arrangements_with_repetition}")
````

### Simple Combination

Simple combination refers to the number of ways to choose $k$ objects from a set of $n$ distinct objects, where order does not matter and repetition is not allowed. The formula is $C(n, k) = \frac{n!}{k!(n-k)!}$.

````python
import math
import itertools

# Example: Simple Combination of 5 items taken 2 at a time
n_comb = 5
k_comb = 2

combinations_count = math.comb(n_comb, k_comb)
print(f"Simple combination of {n_comb} items taken {k_comb} at a time: {combinations_count}")

# Using itertools to list combinations
list_items_comb = ['A', 'B', 'C', 'D', 'E']
combinations_simple = list(itertools.combinations(list_items_comb, k_comb))
print(f"Simple combinations of {list_items_comb} taken {k_comb} at a time: {combinations_simple}")
````

### Combination with Repetition

Combination with repetition refers to the number of ways to choose $k$ objects from a set of $n$ distinct objects, where order does not matter and repetition is allowed. The formula is $C_r(n, k) = C(n+k-1, k) = \frac{(n+k-1)!}{k!(n-1)!}$.

````python
import math
import itertools

# Example: Combination with Repetition of 3 types of fruits (apple, banana, orange)
# choosing 2 fruits (a, b, o)
# n = 3 (types of fruits), k = 2 (fruits to choose)
n_comb_rep = 3
k_comb_rep = 2

# Using the formula C(n+k-1, k)
combinations_with_repetition_count = math.comb(n_comb_rep + k_comb_rep - 1, k_comb_rep)
print(f"Combination with repetition of {n_comb_rep} types of items taken {k_comb_rep} at a time: {combinations_with_repetition_count}")

# Using itertools to list combinations with repetition
items_for_comb_rep = ['apple', 'banana', 'orange']
combinations_with_repetition = list(itertools.combinations_with_replacement(items_for_comb_rep, k_comb_rep))
print(f"Combinations with repetition of {items_for_comb_rep} taken {k_comb_rep} at a time: {combinations_with_repetition}")
````

### Other Useful `itertools` Functions

In addition to combinatorics functions, `itertools` offers other powerful tools for efficiently working with iterators.

*   **`itertools.count(start=0, step=1)`**: Returns an iterator that produces a sequence of integers, starting at `start` and incrementing by `step`. It is an infinite sequence.
    *   **Example:**
    ```python
    import itertools
    # Generate the first 5 numbers from 10 with a step of 2
    for i in itertools.count(10, 2):
        if i > 20: break
        print(i)
    ```

*   **`itertools.cycle(iterable)`**: Returns an iterator that repeats the elements of an `iterable` infinitely.
    *   **Example:**
    ```python
    import itertools
    count = 0
    for item in itertools.cycle(['A', 'B', 'C']):
        if count > 5: break
        print(item)
        count += 1
    ```

*   **`itertools.repeat(object, times=None)`**: Returns an iterator that repeats an `object` a specified number of `times`, or infinitely if `times` is not specified.
    *   **Example:**
    ```python
    import itertools
    for i in itertools.repeat('Hello', 3):
        print(i)
    ```

*   **`itertools.chain(*iterables)`**: Returns an iterator that chains several iterables into a single sequence.
    *   **Example:**
    ```python
    import itertools
    list1 = [1, 2, 3]
    list2 = ['a', 'b']
    chained = list(itertools.chain(list1, list2))
    print(chained)
    ```

*   **`itertools.groupby(iterable, key=None)`**: Returns an iterator that groups consecutive elements of the `iterable` that have the same key (defined by the `key` function). The `iterable` must be **sorted** by the key for `groupby` to work correctly.
    *   **Example:**
    ```python
    import itertools

    data = [('A', 1), ('A', 2), ('B', 3), ('B', 4), ('A', 5)]
    # For groupby to work correctly, the data must be pre-sorted by the key
    data_sorted = sorted(data, key=lambda x: x[0])

    for key, group in itertools.groupby(data_sorted, lambda x: x[0]):
        print(f"Key: {key}, Group: {list(group)}")
    ```
