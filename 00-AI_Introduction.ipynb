# Introdução à Inteligência Artificial (IA)

## 1. Definição de Inteligência Artificial (IA)

A Inteligência Artificial (IA) é um campo da ciência da computação dedicado à criação de sistemas ou máquinas que podem realizar tarefas que normalmente exigiriam inteligência humana. O objetivo principal da IA é capacitar essas máquinas a raciocinar, aprender, resolver problemas, perceber, entender a linguagem e até mesmo criar.

## 2. Básico de História da IA

A ideia de máquinas pensantes remonta à antiguidade, mas o campo da IA como o conhecemos hoje começou a tomar forma em meados do século XX.

*   **Anos 1940-1950:** As primeiras ideias sobre computação e máquinas pensantes surgiram. O teste de Turing (Alan Turing, 1950) propôs um critério para a inteligência da máquina.
*   **1956:** A Conferência de Dartmouth é amplamente considerada o nascimento oficial da IA como um campo de pesquisa. John McCarthy cunhou o termo "inteligência artificial".
*   **Anos 1960-1970:** Período de otimismo e grandes expectativas. Desenvolvimento de sistemas de resolução de problemas (como o General Problem Solver) e primeiros programas de processamento de linguagem natural (como ELIZA).
*   **Anos 1970-1980 (O "Inverno da IA"):** A dificuldade de escalar os primeiros sistemas e o desempenho limitado levaram à desilusão e a cortes de financiamento.
*   **Anos 1980:** O ressurgimento da IA com o sucesso dos sistemas especialistas (programas que encapsulavam o conhecimento de especialistas humanos em um domínio específico).
*   **Anos 1990:** Foco em aprendizagem de máquina, redes neurais e métodos estatísticos. Deep Blue, um computador da IBM, venceu o campeão mundial de xadrez Garry Kasparov em 1997.
*   **Anos 2000-Presente:** Explosão do aprendizado de máquina e, mais tarde, do aprendizado profundo (deep learning), impulsionada por grandes volumes de dados (Big Data), poder computacional (GPUs) e avanços em algoritmos. Conquistas notáveis incluem reconhecimento de fala e imagem, carros autônomos, assistentes virtuais e o desenvolvimento de grandes modelos de linguagem (LLMs).

## 3. Tipos de IA

A IA pode ser categorizada de diversas maneiras, com base em suas capacidades e funcionalidades:

### Por Capacidade:

1.  **IA Fraca (Narrow AI ou ANI - Artificial Narrow Intelligence):** Projetada e treinada para uma tarefa específica (e.g., assistentes de voz, sistemas de recomendação). Não possui consciência ou inteligência generalizada.
2.  **IA Forte (General AI ou AGI - Artificial General Intelligence):** Sistemas de IA que possuem a capacidade de entender, aprender e aplicar inteligência a qualquer tarefa intelectual humana (ainda um objetivo de pesquisa).
3.  **Super IA (Superintelligence ou ASI - Artificial Superintelligence):** Uma IA hipotética que supera a inteligência humana em todas as tarefas e em todas as áreas (puramente especulativo).

### Por Funcionalidade:

*   **Máquinas Reativas (Reactive Machines):** Agem com base na percepção atual; não possuem memória nem aprendem com experiências passadas (e.g., termostatos, Deep Blue).
*   **Memória Limitada (Limited Memory):** Podem usar experiências passadas para informar decisões futuras (e.g., carros autônomos que consideram o tráfego recente).
*   **Teoria da Mente (Theory of Mind):** Uma IA que seria capaz de entender emoções, crenças e intenções (conceito futuro).
*   **Autoconsciente (Self-aware):** O nível mais avançado, onde a máquina teria consciência de si mesma e de sua existência (hipotético).

## 4. Agentes em IA

Em IA, um **agente** é qualquer coisa que pode perceber seu ambiente através de sensores e agir sobre esse ambiente através de atuadores. O objetivo de um agente é maximizar uma medida de desempenho, realizando ações corretas e racionais.

*   **Sensores:** Mecanismos para perceber informações do ambiente.
*   **Atuadores:** Mecanismos para agir sobre o ambiente.
*   **Função do Agente:** Mapeia uma sequência de percepções para uma ação.
*   **Agente Racional:** Aquele que age de forma a atingir o melhor resultado esperado, considerando suas percepções e conhecimento.

## 5. Tipos de Agentes em IA

Os agentes de IA são classificados com base em sua complexidade e operação:

1.  **Agentes de Reflexo Simples (Simple Reflex Agents):** Agem com base na percepção atual usando regras "se-então", ignorando o histórico (e.g., termostato).
2.  **Agentes de Reflexo Baseados em Modelo (Model-Based Reflex Agents):** Mantêm um estado interno (um "modelo") do mundo para tomar decisões com base no estado atual e regras (e.g., carro autônomo).
3.  **Agentes Baseados em Objetivos (Goal-Based Agents):** Têm objetivos e selecionam ações que os levarão a alcançá-los, geralmente envolvendo planejamento (e.g., robô de fábrica).
4.  **Agentes Baseados em Utilidade (Utility-Based Agents):** Consideram não apenas o objetivo, mas a "utilidade" ou "satisfação" do resultado, buscando maximizar a utilidade esperada (e.g., agente de negociação).
5.  **Agentes de Aprendizagem (Learning Agents):** Possuem a capacidade de aprender com suas experiências e melhorar seu desempenho ao longo do tempo (e.g., sistema de recomendação).

## 6. Aprendizado em Inteligência Artificial

Em IA, o **aprendizado** refere-se à capacidade de um sistema de melhorar seu desempenho em uma tarefa específica ao longo do tempo, através da experiência ou da ingestão de dados. Em vez de ser explicitamente programado para cada cenário possível, um sistema de IA aprende a identificar padrões, fazer previsões ou tomar decisões de forma autônoma.

## 7. Tipos de Aprendizado em IA (Machine Learning)

O aprendizado em IA é explorado no campo do **Machine Learning (Aprendizado de Máquina)**. Os principais paradigmas são:

1.  **Aprendizado Supervisionado (Supervised Learning):**
    *   **Conceito:** Treinado com dados "rotulados" (entradas com saídas desejadas). O algoritmo aprende a mapear entradas para saídas.
    *   **Exemplos:** Classificação (e.g., spam/não-spam), Regressão (e.g., preço de casas).
    *   **Algoritmos:** Regressão Linear/Logística, SVMs, Árvores de Decisão, Redes Neurais.

2.  **Aprendizado Não Supervisionado (Unsupervised Learning):**
    *   **Conceito:** Treinado com dados "não rotulados". O algoritmo encontra estruturas, padrões ou relações ocultas nos dados.
    *   **Exemplos:** Agrupamento (clustering, e.g., segmentação de clientes), Redução de Dimensionalidade (e.g., PCA).
    *   **Algoritmos:** K-Means, PCA, Autoencoders.

3.  **Aprendizado por Reforço (Reinforcement Learning - RL):**
    *   **Conceito:** O sistema (agente) aprende a tomar decisões sequenciais em um ambiente dinâmico para maximizar uma "recompensa" cumulativa, por tentativa e erro.
    *   **Exemplos:** Jogos (AlphaGo), Robótica, Navegação autônoma.
    *   **Algoritmos:** Q-Learning, Deep Q-Networks (DQN).

4.  **Aprendizado Semi-supervisionado (Semi-supervised Learning):**
    *   **Conceito:** Combina dados rotulados e não rotulados para treinamento, útil quando a rotulagem é cara.
    *   **Exemplos:** Reconhecimento facial (com algumas imagens rotuladas).

5.  **Aprendizado Ativo (Active Learning):**
    *   **Conceito:** O algoritmo seleciona ativamente os exemplos mais informativos para serem rotulados por um especialista humano.
    *   **Exemplos:** Medicina (onde rotular dados requer conhecimento especializado).
